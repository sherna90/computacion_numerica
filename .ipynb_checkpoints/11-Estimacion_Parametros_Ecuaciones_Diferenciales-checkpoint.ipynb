# Ecuaciones Diferenciales Ordinarias

Una ecuación diferencial es una ecuación que expresa la relación entre una función y sus derivadas.

Cuando la función $f(x)$ depende únicamente de una variable $x \in \mathbb R$ definida en una dimensión, entonces la ecuación se dice **ordinaria**.

$$F\left(x, f(x), \frac{df(x)}{dx}, \frac{d^2f(x)}{dx^2}, \frac{d^3f(x)}{dx^3}, \ldots, \frac{d^{n-1}f(x)}{dx^{n-1}}\right) = \frac{d^{n}f(x)}{dx^{n}}$$

Donde $n$ es el orden de la EDO.

## EDO de primer orden

Por ejemplo, supongamos que contamos con la siguiente EDO:

$$\begin{aligned}
\frac{\mathrm{d}x}{\mathrm{d}t} &= (\alpha - \beta y(t))x(t), \
\frac{\mathrm{d}y}{\mathrm{d}t} &= (\delta x(t) - \gamma)y(t)
\end{aligned}$$

Donde $x$ es la presa e $y$ es la población de depredadores. 

```julia
# Define Lotka-Volterra model.
function lotka_volterra(du, u, p, t)
    # Model parameters.
    α, β, γ, δ = p
    # Current state.
    x, y = u

    # Evaluate differential equations.
    du[1] = (α - β * y) * x # prey
    du[2] = (δ * x - γ) * y # predator

    return nothing
end


```

```julia
# Imprimir los parámetros iniciales
p = [1.5, 1.0, 3.0, 1.0];
println("Parámetros iniciales:")
println("α (tasa de crecimiento de la liebre): ", p[1])
println("β (tasa de depredación del lince sobre la liebre): ", p[2])
println("γ (tasa de mortalidad del lince): ", p[3])
println("δ (tasa de crecimiento del lince debido a la depredación): ", p[4])
```

```julia
using DifferentialEquations;

# Define initial-value problem.
u0 = [1.0, 1.0];
p = [1.5, 1.0, 3.0, 1.0];
tspan = (0.0, 10.0);
prob = ODEProblem(lotka_volterra, u0, tspan, p);
```

```julia
using Plots;

# Plot simulation.
plot(solve(prob, Euler(),dt=0.01))
```

Ahora, hacemos la estimación de los parámetros $\alpha, \beta, \gamma, \delta$ 

```julia
using CSV;
using Downloads;

url = "https://raw.githubusercontent.com/stan-dev/example-models/master/knitr/lotka-volterra/hudson-bay-lynx-hare.csv"
http_response = Downloads.download(url);
file = CSV.File(http_response,header=["Year", "Lynx", "Hare"], skipto=4);
df = DataFrame(file);
year = df.Year;
hare = df.Hare;
lynx = df.Lynx;
```

```julia
plot(year,hare,label="Hare")
plot!(year,lynx,label="Lynx")
```

```julia
u0=zeros(2)
t0,u0=year[1],[hare[1],lynx[1]]

tspan=year[1],year[end]
p = [1.5, 1.0, 3.0, 1.0];
prob = ODEProblem(lotka_volterra, u0, tspan, p);
```

```julia
using Printf;
sol = solve(prob,Euler(),dt=0.001,saveat=year);
loss = sum((sol(year[i])[1] - hare[i])^2 + (sol(year[i])[2] - lynx[i])^2 for i in 1:length(year))/length(year)
@printf("Error Cuadratico Medio : %e ",loss)
```

```julia
function loss_function(p)
    prob = ODEProblem(lotka_volterra, u0, tspan,p);
    sol = solve(prob,Euler(),dt=0.001,saveat=1);
    loss = sum((sol(year[i])[1] - hare[i])^2 + (sol(year[i])[2] - lynx[i])^2 for i in 1:length(year))/length(year)
    return loss
end;
```

```julia
using Optim;

lower = [0.01, 0.01, 0.01, 0.01];
upper = [4.0, 4.0, 4.0, 4.0];
initial = [1.5, 1.0, 3.0, 1.0];
```

```julia
result = optimize(loss_function,initial, NelderMead())
```

```julia
p_estimated = result.minimizer

# Imprimir los parámetros estimados
println("Parámetros estimados:")
println("α (tasa de crecimiento de la liebre): ", p_estimated[1])
println("β (tasa de depredación del lince sobre la liebre): ", p_estimated[2])
println("γ (tasa de mortalidad del lince): ", p_estimated[3])
println("δ (tasa de crecimiento del lince debido a la depredación): ", p_estimated[4])
```

```julia
result_sa = optimize(loss_function,lower,upper,initial,SAMIN(rt=0.95), Optim.Options(iterations=10^6))
```

```julia
p_estimated_sa= result_sa.minimizer

# Imprimir los parámetros estimados
@printf("Parámetros estimados:")
@printf("α (tasa de crecimiento de la liebre): %.3f\n", p_estimated_sa[1])
@printf("β (tasa de depredación del lince sobre la liebre): %.3f\n", p_estimated_sa[2])
@printf("γ (tasa de mortalidad del lince): %.3f\n", p_estimated_sa[3])
@printf("δ (tasa de crecimiento del lince debido a la depredación): %.3f\n", p_estimated_sa[4])
```

```julia
prob = ODEProblem(lotka_volterra, u0, tspan, p_estimated);
sol = solve(prob,Euler(),dt=0.001,saveat=year);
plot(sol)
```
