En esta tarea su grupo debe ajustar un sistema de ecuaciones no-lineales para la función de Lorenz. 

Cada grupo debe etiquetarse con un numero de grupo (id_grupo). Este numero se usa para generar los datos que serán usados para formar el sistema de ecuaciones.  

```julia
id_grupo=12
```

```julia
using LinearAlgebra
using Plots

f(x,α,β,γ)=α/(1.0+((γ-x)/β)^2)
```

```julia
x=0:0.1:2;
y=f.(x,2.,1.,1.);

plot(x,y)
```

```julia
using Random

rng = MersenneTwister(id_grupo);

num_points=10
X=rand(rng,x,num_points);
Y=f.(X,2.,1.,1.)+randn(rng,num_points)*0.1;
```

```julia
plot(X,Y,seriestype=:scatter)
```
