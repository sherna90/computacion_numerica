## 1. Biseccion 

Utilice el metodo de la biseccion para encontrar los ceros de la funcion $f(x)=2x*cos(\pi*x)-exp({x-1})$ definida en el intervalo $[1 ,2]$. Encontrar el numero de iteraciones $n$ requerido para encontrar un error $\varepsilon < 10^-3$ mediante el criterio $| f(x_n) |<\varepsilon$

```julia
using Plots 

f(x)=2x*cos(π*x)-exp(x-1)

x=0-2.0:0.1:2.0
y=f.(x)

plot(x,y)
```

```julia
using Roots

tracker = Roots.Tracks()
x0=find_zero(f,(1.0,2.0), Bisection(),tracks=tracker,atol=1e-3)
```

```julia
tracker1 = Roots.Tracks()
x1=find_zero(f,(-2.0,-1.0), Bisection(),tracks=tracker1,rtol=1e-3)
```

```julia
tracker
```

```julia
tracker1
```

```julia
x0
```

```julia
round(x0,digits=3)
```

## 2. Newton-Raphson 
Utilice el metodo de Newton-Raphson para encontrar los ceros de la funci\'on $f(x)=-x^3-\text{cos}(x)$ definida en el intervalo $[-\pi/2 ,\pi/2]$. Encontrar el número de ceros y la cantidad de iteraciones $n$ requeridas para encontrar un error $\varepsilon < 10^{-6}$ mediante el criterio $| f(x_n) |<\varepsilon$

```julia
tracker = Roots.Tracks()

f(x)=-x^3-cos(x)
df(x)=-3x^2+sin(x)

x=-π/2:0.1:π/2

y=f.(x)

plot(x,y)
```

```julia
x0=find_zero((f,df),-1.0, Roots.Newton(),tracks=tracker,atol=1e-6)
```

```julia
tracker
```

```julia
f(x0)
```

```julia
round(x0,digits=6)
```

```julia
abs(f(-0.865475))
```

```julia
abs(f(-0.865473))
```

## 3. Secante

Utilice el metodo de la secante para encontrar los ceros de la funcion $f(x)=x^3+3x^2-1$ definida en el intervalo $[-3,-2]$. Encontrar la raiz $x_n$ de modo que el error sea $\varepsilon < 10^-6$ mediante el criterio $| f(x_n) |< \varepsilon$

```julia
tracker = Roots.Tracks()

f(x)=x^3+3x^2-1

x=-3.0:0.1:-2.0

y=f.(x);
```

```julia
plot(x,y)
```

```julia
x0=find_zero(f,-3.0, Roots.Secant(),tracks=tracker,atol=1e-6)
```

```julia
tracker
```

```julia
abs(f(x0))
```

```julia
abs(f(round(x0,digits=6)))
```
