1. Utilice el metodo de la bisecci\'on para encontrar los ceros de la funcion $f(x)=x-3^{-x}$ definida en el intervalo $[0 ,1]$. Encontrar el numero de iteraciones $n$ requerido para encontrar un error $\varepsilon < 10^-5$ mediante el criterio $| f(x_n) |<\varepsilon$

```julia
using Plots 

f(x)=x-3^-x

x=0.0:0.1:1.0
y=f.(x)

plot(x,y)
```

```julia
using Roots

tracker = Roots.Tracks()
x0=find_zero(f,(0.0,1.0), Bisection(),tracks=tracker,atol=1e-5)
```

```julia
tracker
```

4. Utilice el metodo de Newton-Raphson para encontrar los ceros de la funci\'on $f(x)=x^3-5x^2+5$ definida en el intervalo $[-2 ,2]$. Encontrar el n\'umero de ceros y la cantidad de iteraciones $n$ requeridas para encontrar un error $\varepsilon < 10^-6$ mediante el criterio $| f(x_n) |<\varepsilon$

```julia
tracker = Roots.Tracks()

f(x)=x^3-5x^2+5
df(x)=3x^2-10x

x=-2.0:0.1:2.0

y=f.(x)

plot(x,y)
```

```julia
x0=find_zero((f,df),-1.0, Roots.Newton(),tracks=tracker,atol=1e-6)
```

```julia
tracker
```
